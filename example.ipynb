```python
import obspy
import mkwin
```

```python
tr = obspy.read('http://examples.obspy.org/RJOB_061005_072159.ehz.new')[0]
tr.plot()
```

```python
print(tr.data)
st = tr.stats["starttime"].datetime
print(st)
fs = tr.stats["sampling_rate"]
print(fs)
mkwin.mkwin(
    tr.data,
    sampling_freq = int(fs),
    startdatetime = st,
    savedir = "ret",
    savename = "example.win",
    overwrite = True,
)
print("END")
```
